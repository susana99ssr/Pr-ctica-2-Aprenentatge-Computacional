Los atributos "Location", "WindGustDir", "WindDir9am", "WindDir3pm", "RainToday" y "RainTomorrow" son las variables categóricas. Los demás atributos son numéricos. El "RainTomorrow" será el atributo objetivo, porque requiere una predicción de una de las dos categorías ("Yes" o "No").

```python
categoricas = ["Location", "WindGustDir", "WindDir9am", "WindDir3pm", "RainToday", "RainTomorrow"]
```

```python
data[categoricas] = data[categoricas].astype("category")
data[categoricas].describe()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data["Date"].describe(datetime_is_numeric=True)
```

El atributo objetivo no está blanceado (tiene 110316 entradas negativas, que es mucho más que las positivas)
