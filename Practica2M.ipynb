## Preprocessing

Como hemos visto en el apartado anterior, los datos numéricos no están estandarizados. Hay que estandarizarlos para procesar los valores en el mismo rango (promedio = 0, desviación estándar = 1). Así el procesamiento será más rápido.

Algunos atributos tienen valores nulos. Los vamos a tratar de maneras diferentes, más adientes para cada tipo de dato en concreto.

Los valores nulos de atributos numéricos serán sustituidos por valores promedios de dichos atributos para cada par (Mes, Ciudad). Primero, construimos la tabla agrupada por ciudades y fefchas con valores promedios de atributos numéricos. Aprovechamos de que no tenemos valores nulos para localidades y fechas.

```python
from sklearn.preprocessing import LabelEncoder
mean_features = ["Evaporation","MaxTemp","Rainfall","MinTemp","Sunshine","WindGustSpeed","WindSpeed9am","WindSpeed3pm",
                "Humidity9am","Humidity3pm","Pressure9am","Pressure3pm","Cloud9am","Cloud3pm","Temp9am","Temp3pm"]
mean_table = data.groupby(["Month", "Location"], as_index = False)[mean_features].mean()
mean_table.reset_index()
le = LabelEncoder()
mean_table["Location"] = le.fit_transform(mean_table["Location"])
data["Location"] = le.fit_transform(data["Location"])
mean_table
```

Como se ve, esta tabla también tiene valores nulos. Como esta segunda tabla tiene pocas filas, podemos utilizar el KNN Imputer.

```python
from sklearn.impute import KNNImputer
knni = KNNImputer()
mean_table = pd.DataFrame(knni.fit_transform(mean_table),columns = mean_table.columns)
mean_table.info()
```

```python
for feature in mean_features:
    for month in data["Month"].unique():
        for location in data["Location"].unique():
            mask = (data["Location"] == location) & (data["Month"] == month)
            mask_mean = (mean_table["Location"] == location) & (mean_table["Month"] == month)
            val = mean_table.loc[mask_mean, feature].item()
            data.loc[mask, feature] = data.loc[mask,feature].fillna(val)
```

## Model selection

Considerando el tamaño de nustra base de datos se puede eligir los siguientes modelos: Kernel SVM, Multilayer Perceptron, Random Forest, Gradient Boosting. Suponemos que serán más precisos. Pero tambien se puede considerar tambien los árboles de decisión, regresión logística y Naïve Bayes, porque se ajustarán más rápido.

Hemos decidido usar el RBF kernel, porque suponemos que la distribución de clases de nuestra variable objetivo no puede ser lineal (la probabilidad de la lluvia está compuesta por muchas circunstancias que no dependen una de otra).

Los modelos ensemble seguramente serán precisos, pero tardarán bastante, ya que se usan mas para datasets más pequeños

## Crossvalidation

La cross-validación de los resultados es la parte más importante en la construcción de un buen modelo, que puede dar buenos resultados en el conjnunto de entrenamiento, pero fracasará en casos no vistos. Cross-validación no ayudará a evitar el overfitting y el underfitting.

La mejor estrategia para separar el dataset en conjuntos de entrenamiento y validación es guardar más datos en el train set, porque así utilizamos casi todos los datos disponibles. Pero no es posible evitar el test-set, porque sin él la probabilidad de overfitting es mucho mayor. Nuestra base de datos es bastante grande para poder probar separar el dataset en 90/10 de train y test, respectivamente.

También, se puede usar la estrategia K-Fold con mayor K posible y con mayor número de repeticiones. De esta manera el tipo más interesante puede ser el Repeated Stratified K-Fold. Pero tardará bastante. Habrá que probar diferentes tipos.

En nuestro caso seguramente no podemos utilizar la estrategia Leave-One-Out, ya que así tendrémos más de 100 mil train sets, que es demasiado costoso.

## Metric Analysis

Para evaluar la calidad de nuestro clasificador no podemos utilizar accuracy, porque los datos no están balanceados. Con tantos valores negativos lo que necesitamos es F1-Score, porque es una combinación de Precision y Recall.
Tambien necesitamos AUC.

La Precisión-Recall Curve y la ROC Curve serán igualmente relevantes para nustra base de datos, porque nos dan toda la información desde puntos de vista diferentes. 
PRC es la curva que muestra la relación entre Precision y Recall. ROC muestra la relación entre False Positive Rate y True Positive Rate.
