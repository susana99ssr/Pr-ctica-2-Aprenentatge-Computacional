## Preprocessing

Como hemos visto en el apartado anterior, los datos num√©ricos no est√°n estandarizados. Hay que estandarizarlos para procesar los valores en el mismo rango (promedio = 0, desviaci√≥n est√°ndar = 1). As√≠ el procesamiento ser√° m√°s r√°pido.

Algunos atributos tienen valores nulos. Los vamos a tratar de maneras diferentes, m√°s adientes para cada tipo de dato en concreto.

Los valores nulos de atributos num√©ricos ser√°n sustituidos por valores promedios de dichos atributos para cada par (Mes, Ciudad). Primero, construimos la tabla agrupada por ciudades y fefchas con valores promedios de atributos num√©ricos. Aprovechamos de que no tenemos valores nulos para localidades y fechas.

```python
from sklearn.preprocessing import LabelEncoder
mean_features = ["Evaporation","MaxTemp","Rainfall","MinTemp","Sunshine","WindGustSpeed","WindSpeed9am","WindSpeed3pm",
                "Humidity9am","Humidity3pm","Pressure9am","Pressure3pm","Cloud9am","Cloud3pm","Temp9am","Temp3pm"]
mean_table = data.groupby(["Month", "Location"], as_index = False)[mean_features].mean()
mean_table.reset_index()
le = LabelEncoder()
mean_table["Location"] = le.fit_transform(mean_table["Location"])
data["Location"] = le.fit_transform(data["Location"])
mean_table
```

Como se ve, esta tabla tambi√©n tiene valores nulos. Como esta segunda tabla tiene pocas filas, podemos utilizar el KNN Imputer.

```python
from sklearn.impute import KNNImputer
knni = KNNImputer()
mean_table = pd.DataFrame(knni.fit_transform(mean_table),columns = mean_table.columns)
mean_table.info()
```

```python
for feature in mean_features:
    for month in data["Month"].unique():
        for location in data["Location"].unique():
            mask = (data["Location"] == location) & (data["Month"] == month)
            mask_mean = (mean_table["Location"] == location) & (mean_table["Month"] == month)
            val = mean_table.loc[mask_mean, feature].item()
            data.loc[mask, feature] = data.loc[mask,feature].fillna(val)
```

## Crossvalidation

La cross-validaci√≥n de los resultados es la parte m√°s importante en la construcci√≥n de un buen modelo, que puede dar buenos resultados en el conjnunto de entrenamiento, pero fracasar√° en casos no vistos. Cross-validaci√≥n no ayudar√° a evitar el overfitting y el underfitting.

La mejor estrategia para separar el dataset en conjuntos de entrenamiento y validaci√≥n es guardar m√°s datos en el train set, porque as√≠ utilizamos casi todos los datos disponibles. Pero no es posible evitar el test-set, porque sin √©l la probabilidad de overfitting es mucho mayor. Nuestra base de datos es bastante grande para poder probar separar el dataset en 90/10 de train y test, respectivamente.

Tambi√©n, se puede usar la estrategia K-Fold con mayor K posible y con mayor n√∫mero de repeticiones. De esta manera el tipo m√°s interesante puede ser el Repeated Stratified K-Fold. Pero tardar√° bastante. Habr√° que probar diferentes tipos.

En nuestro caso seguramente no podemos utilizar la estrategia Leave-One-Out, ya que as√≠ tendr√©mos m√°s de 100 mil train sets, que es demasiado costoso.

## Metric Analysis

Para evaluar la calidad de nuestro clasificador no podemos utilizar accuracy, porque los datos no est√°n balanceados. Con tantos valores negativos lo que necesitamos es F1-Score, porque es una combinaci√≥n de Precision y Recall.
Tambien necesitamos AUC.

La Precisi√≥n-Recall Curve y la ROC Curve ser√°n igualmente relevantes para nustra base de datos, porque nos dan toda la informaci√≥n desde puntos de vista diferentes. 
PRC es la curva que muestra la relaci√≥n entre Precision y Recall. ROC muestra la relaci√≥n entre False Positive Rate y True Positive Rate.
